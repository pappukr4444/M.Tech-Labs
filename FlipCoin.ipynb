```python
#head=1
#tail=0
r=0
from random import sample
import matplotlib.pyplot as plt
import numpy as np


head=[0]*10
tail=[0]*10
m=[0]*1001

temp=2

a=0
b=0
def equ():
    global a
    global b
    global temp
    if temp==0:
        a=a+1
    else:
        b=b+1
        
def save():
    
    global a
    global b
    global temp
    if temp==0:
        tail[a]=tail[a]+1
    else:
        head[b]=head[b]+1
    
            
def overflow():
    global a
    global b
    
    if a<10 or b<10:
        return
    elif a==10 or b==10:
        save()
        set()
        return
def set():
    global a
    global b
    a=0
    b=0

    
    
for i in range(10):
    
    l=0
    temp=2
    
    r=r+1
    head=[0]*15
    tail=[0]*15
    set()
    print("")
    print("")
    print("Iteration number -- "+str(i+1))
    print("")
    for j in range(1000):
        
        n=(sample(range(1,1000),1))
        m[j]=int(n[0]%2)
    for k in range (1000):
        if temp==2:
            temp=m[k]
            equ()
        elif temp==m[k]:
            overflow()
            equ()
        else:
            save()
            set()
            temp=m[k]
            equ()
    for l in range(10):
        t=l+1
        
        print(str(t) +" consecutive heads ="+str(head[t])+" and  "+str(t)+" consecutive tails ="+str(tail[t]))
  
    fig1 = plt.figure()
    #ax1 = f1.add_subplot(111)
    #plt.set_xlabel('Value')
    plt.xlabel('Value')
    plt.ylabel('Frequency')
    
    #ax1.set_ylabel('Frequency')
    plt.hist(head, 10, range=[0, 12], facecolor='green', align='mid')
    plt.ylim(0, 10)
    
    
    fig2 = plt.figure()
    plt.xlabel('Value')
    plt.ylabel('Frequency')
    #ax2.set_xlabel('Value')
    #ax2.set_ylabel('Frequency')
    plt.hist(tail, 10, range=[0, 12], facecolor='blue', align='mid')
    plt.ylim(0, 10)


    plt.show()
```
